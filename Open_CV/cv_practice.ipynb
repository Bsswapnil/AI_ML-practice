```python
import matplotlib.pyplot as plt
import numpy as np
```

```python


# Create sample images
gray_image = np.random.rand(10, 10)  # Random grayscale image
rgb_image = np.random.rand(10, 10, 3)  # Random RGB image

# Create subplots
plt.figure(figsize=(8, 4))

# Grayscale image
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# RGB image
plt.subplot(1, 2, 2)
plt.imshow(rgb_image)
plt.title('RGB Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
kernel = np.ones((5,5),np.uint8)

# img_chess_path = os.path.join(os.path.dirname(__file__), 'images', 'chess.png')
img_chess_path = "/home/swapnil/ACL/OpenCV/OpenCV-Python-Tutorials-and-Projects/Resources/chess.jpg"
img_chess_openCV = cv2.imread(img_chess_path)
img_Gray = cv2.cvtColor(img_chess_openCV,cv2.COLOR_BGR2GRAY)
img_Blur = cv2.GaussianBlur(img_Gray,(7,7),0)
img_Canny = cv2.Canny(img_Blur,100,200)
img_Dilation = cv2.dilate(img_Canny, kernel, iterations = 10)
img_Eroded = cv2.erode(img_Dilation, kernel, iterations=2)

plt.subplot(2,3,1)
plt.imshow(img_chess_openCV)
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,3,2)
plt.imshow(img_Gray,cmap='gray')
plt.title('Gray Image')
plt.axis('off')

plt.subplot(2,3,3)
plt.imshow(img_Blur,cmap='gray')
plt.title('Blurred Image')
plt.axis("off")

plt.subplot(2,3,4)
plt.imshow(img_Canny,cmap='gray')
plt.title('Canny Image')
plt.axis("off")

plt.subplot(2,3,5)
plt.imshow(img_Dilation,cmap='gray')
plt.title('Dilated Image')
plt.axis("off")

plt.subplot(2,3,6)
plt.imshow(img_Eroded,cmap='gray')
plt.title('Eroded Image')
plt.axis("off")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load an image
img = cv2.imread("road.jpg")  # Replace with the correct path to your image

# Resize the image
imgResize = cv2.resize(img, (600, 400))

# Crop a portion of the image
imgCropped = img[300:540, 430:480]

# Resize the cropped image back to the original size
imgCropResize = cv2.resize(imgCropped, (img.shape[1], img.shape[0]))

# Convert images to RGB for Matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imgResize = cv2.cvtColor(imgResize, cv2.COLOR_BGR2RGB)
imgCropped = cv2.cvtColor(imgCropped, cv2.COLOR_BGR2RGB)
imgCropResize = cv2.cvtColor(imgCropResize, cv2.COLOR_BGR2RGB)

# Display images
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(img)
plt.axis("off")

plt.subplot(2, 2, 2)
plt.title("Resized Image")
plt.imshow(imgResize)
plt.axis("off")

plt.subplot(2, 2, 3)
plt.title("Cropped Image")
plt.imshow(imgCropped)
plt.axis("off")

plt.subplot(2, 2, 4)
plt.title("Cropped & Resized Image")
plt.imshow(imgCropResize)
plt.axis("off")

plt.tight_layout()
plt.show()
```
